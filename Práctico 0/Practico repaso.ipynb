```python
# Funciones
#Manera de compactar y reutilizar el código. Es de utilidad cuando necesitamos realizar la misma acción muchas veces
#En Python no es necesario declarar el tipo de dato de entrada o salida, pudiendo o no retornar algo.


def divide(dividendo, divisor):
    """
    Función que devuelve el cociente y el resto de una división.

    Parameters:
    - dividendo (int): El número que se va a dividir.
    - divisor (int): El número por el cual se va a dividir.

    Returns:
    tuple: Una tupla que contiene el cociente y el resto.
    """
    cociente = dividendo // divisor
    resto = dividendo % divisor
    return cociente, resto


c, r = divide(10,3)
print(c, r)
```

```python
#Condicionales if
if r == 0:
  print('división exacta')
else:
  print('división entera')
```

```python
#Ciclos
d = 10
print('Dividendo: ', d)
for divisor in range(1, 10): #secuencia de numeros enteros con una distancia (steps) específica. Por defecto 1
  c, r = divide(d,divisor)
  if r == 0:
    print('división exacta. Divisor: ', divisor)
  else:
    mensaje = f'División entera. Resto: {r}, Divisor: {divisor}'
    print(mensaje)
```

```python
#complejizamos la función
def divisores_exactos(dividendo, max_divisor):
    divisores_exactos_ = []
    print('Dividendo:', dividendo)
    for i in range(1, max_divisor):
        c, r = divide(dividendo, i)
        if r == 0:
            divisores_exactos_.append(i)
            print(f'División exacta. Divisor: {i}')
    return divisores_exactos_

# Ejemplo de uso de la función
dividendo_ejemplo = 20
max_divisor_ejemplo = 10
divisores_encontrados = divisores_exactos(dividendo_ejemplo, max_divisor_ejemplo)
print('Divisores exactos encontrados:', divisores_encontrados)

```
